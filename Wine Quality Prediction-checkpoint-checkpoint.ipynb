# Data Preprocessing

```python
X = df.iloc[:,:-1]
X
```

```python
y = df.iloc[:,-1]
y
```

# Train the Model

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

# Using DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_pred1 = dtc.predict(X_test)
y_pred1
```

```python
from sklearn.metrics import accuracy_score
acc_Sc1 = accuracy_score(y_pred1,y_test)
acc_Sc1
```

# Using RidgeClassifier

```python
from sklearn.linear_model import RidgeClassifier
rc = RidgeClassifier()
rc.fit(X_train, y_train)
```

```python
y_pred2 = rc.predict(X_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score
acc_Sc2 = accuracy_score(y_pred2,y_test)
acc_Sc2
```

# Using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred3= rfc.predict(X_test)
y_pred3
```

```python
from sklearn.metrics import accuracy_score
acc_Sc3 = accuracy_score(y_pred3,y_test)
acc_Sc3
```

# Using Naive_Bayes(GaussianNB)

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(X_train, y_train)
```

```python
y_pred4 = NB.predict(X_test)
y_pred4
```

```python
from sklearn.metrics import accuracy_score
acc_Sc4 = accuracy_score(y_pred4,y_test)
acc_Sc4
```

# Using KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
KNc = KNeighborsClassifier()
KNc.fit(X_train, y_train)
```

```python
y_pred5 = KNc.predict(X_test)
y_pred5
```

```python
from sklearn.metrics import accuracy_score
acc_Sc5 = accuracy_score(y_pred5,y_test)
acc_Sc5
```

# Using SupportVectorClassifier(SVC)

```python
from sklearn.svm import SVC
svmc = SVC()
svmc.fit(X_train, y_train)
```

```python
y_pred6 = svmc.predict(X_test)
y_pred6
```

```python
from sklearn.metrics import accuracy_score
acc_Sc6 = accuracy_score(y_pred6,y_test)
acc_Sc6
```

```python
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_pred6,y_test)
cm1
```

```python
plt.figure(figsize=(3,3))
sns.heatmap(cm1, annot=True, cmap='coolwarm',fmt=".2F",cbar=False)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
print(f"Acc_score in DecisionTreeClassifier is  {acc_Sc1}")
print(f"Acc_score in RidgeClassifier is         {acc_Sc2}")
print(f"Acc_score in RandomForestClassifier is  {acc_Sc3}")
print(f"Acc_score in GaussianNB  is             {acc_Sc4}")
print(f"Acc_score in KNeighborsClassifier is    {acc_Sc5}")
print(f"Acc_score in SupportVectorClassifier is {acc_Sc6}")
```
